# Using Pandas to Prep Amazon Review Dataset 
### In order to utilize NLP we will need to make the following modifications & assumptions
#### Assumptions 
####     4 & 5 Start reviews = Most likely a positive review 
####      3 Star reviews = Most likely neither a positive or negitive review, or something that has both 
####      2 & 1 Star Reviews = Most likely a negitive review 

#### Mofications and Cleaning Required 
#### 1) Review Null values from text and summary columns 
####  2) Add in if the review is Positive or Neigitve based off of our star rating assumption 
####  3) Remove unneccary columns 
#### 4) Create seperate dataframs for Text and Summary

```python
import pandas as pd 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
##Loading the Data 

reviews= pd.read_csv('https://data-bootcamp-x399.s3.us-east-2.amazonaws.com/Reviews.csv')

reviews=pd.DataFrame(reviews)
reviews.head()
```

```python
#Adding star rating Potive vs Negitive Assumption 
reviews_sent = reviews[reviews['Score'] != 3]
reviews_sent['sentiment'] = reviews_sent['Score'].apply(lambda rating : 'positive' if rating > 3 else 'negative')
reviews_sent.head()
                     
                     
```

```python
# Check results to ensure only Positive and Negative values
reviews_sent.sentiment.unique()
```

```python
#Cleaning the data for our NLP review

reviews_sum_nlp = reviews_sent.drop(['Id', 'ProductId','UserId','ProfileName','HelpfulnessNumerator','HelpfulnessDenominator','Score','Time','Text'], axis=1)
reviews_sum_nlp.rename(columns={'Summary':'text'}, inplace=True)
reviews_sum_nlp.rename(columns={'sentiment':'class'}, inplace=True)
reviews_sum_nlp.head()
```

```python
reviews_sum_nlp.to_csv('nlp_summary.csv')
```
